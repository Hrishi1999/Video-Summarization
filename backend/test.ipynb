```python
from matplotlib.pyplot import figure, imshow, axis
from matplotlib.image import imread

def showImagesHorizontally(list_of_files):
    fig = figure(figsize=(20,20))
    number_of_files = len(list_of_files)
    for i in range(number_of_files):
        a=fig.add_subplot(1, number_of_files,i+1)
        image = imread("extracted_frames/" + list_of_files[i])
        imshow(image,cmap='Greys_r')
        axis('off')
    fig.savefig('x.png')
```

```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir("extracted_frames") if isfile(join("extracted_frames", f))]

# tfiles = ["output/cluster0/frame90.jpg", "output/cluster1/frame230.jpg"]
showImagesHorizontally(onlyfiles)
```
